Create Table by Copy Python sqlalchemy

```python
import pandas as pd
```

```python
f1 = 'C:\\Users\\Admin\\Downloads\\original\\\data2\\goods_customer.csv'
d1 = 'goods_customer'
f2 = 'C:\\Users\\Admin\\Downloads\\original\\\data2\\goods.csv'
d2 = 'goods'
f3 = 'C:\\Users\\Admin\\Downloads\\original\\\data2\\goods_promotion.csv'
d3 = 'goods_promotion'
f4 = 'C:\\Users\\Admin\\Downloads\\original\\\data2\\goods_sell.csv'
d4 = 'goods_sell'
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://*****:************@localhost:5432/postgres')
```

```python
datalist = [d1, d2, d3, d4]
filelist = [f1, f2, f3, f4]
```

```python
for i in range(10):
    d = datalist[i]
    f = filelist[i]
    df = pd.read_csv(f)
    print('-----data read from csv into DF' + d + '------')
    print(df.head())
    sql = 'DROP TABLE IF EXISTS public.' + d + "CASCADE;"
    engine.execute(sql)
    df.to_sql(d,engine)
    print('----- data converted into database' + d + '------')
    sql = 'SELECT * from public.' + d
    data_db = pd.read_sql_query((sql), engine)
    print(data_db.head())
    
print("complete data injestion!")
```
